```python
import numpy as np
import pandas as pd
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
df = movies.merge(credits,how ='inner', on='title')
```

```python
df.head(1)
```

```python
#genres
#id
#keywords
#original_language
#overview
#title
#cast
#crew
new_df = df[['id','title','genres','keywords','overview','cast','crew']]
```

```python
new_df
```

```python
new_df.isnull().sum()
```

```python
new_df.dropna(inplace=True)
```

```python
import ast
def extract_genre(genres):
    genres_list = ast.literal_eval(genres)
    l=[]
    for dicts in genres_list:
        l.append(dicts['name'])
    return l
    
```

```python
new_df['genres'] = new_df['genres'].apply(extract_genre)
```

```python
new_df
```

```python
new_df['keywords'][0]
```

```python
new_df['keywords'] = new_df['keywords'].apply(extract_genre)
```

```python
new_df
```

```python
new_df['overview'] = new_df['overview'].apply(lambda x: x.split(' '))
```

```python
new_df['crew'][0]
```

```python

def extract_cast(cast):
    cast_list = ast.literal_eval(cast)
    l=[]
    for ele in cast_list[0:3]:
        l.append(ele['character'])
    return l
```

```python

def extract_director(crew):
    crew_list = ast.literal_eval(crew)
    l=[]
    for ele in crew_list:
        if ele['job'] == 'Director':
            l.append(ele['name'])
    return l
```

```python
new_df['cast'] = new_df['cast'].apply(extract_cast)
```

```python
new_df['crew'] = new_df['crew'].apply(extract_director)
```

```python
new_df
```

```python
new_df['genres'] = new_df['genres'].apply(lambda x:[i.replace(' ','') for i in x])
new_df['keywords'] = new_df['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
new_df['overview'] = new_df['overview'].apply(lambda x:[i.replace(' ','') for i in x])
new_df['cast'] = new_df['cast'].apply(lambda x:[i.replace(' ','') for i in x])
new_df['crew'] = new_df['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
new_df
```

```python
new_df['Tags'] = new_df['genres'] + new_df['keywords'] + new_df['overview'] + new_df['cast'] + new_df['crew']
```

```python
new_df
```

```python
new_df = new_df[['id','title','Tags']]
```

```python
new_df['Tags'] = new_df['Tags'].apply(lambda x: ' '.join(x))
```

```python
new_df['Tags'] = new_df['Tags'].apply(lambda x: x.lower())
```

```python
new_df
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
    
```

```python
new_df['Tags'] = new_df['Tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features = 5000, stop_words = 'english')
vectors = vectorizer.fit_transform(new_df['Tags']).toarray()
vectorizer.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x: x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
